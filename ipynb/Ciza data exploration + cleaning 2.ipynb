## sextractor flag notations

1 The object has neighbours, bright and close enough to significantly bias the MAG AUTO  
photometry13, or bad pixels (more than 10% of the integrated area affected),  
2 The object was originally blended with another one,  
4 At least one pixel of the object is saturated (or very close to),  
8 The object is truncated (too close to an image boundary),  
16 Object’s aperture data are incomplete or corrupted,  
32 Object’s isophotal data are incomplete or corrupted,  
64 A memory overflow occurred during deblending,  
128 A memory overflow occurred during extraction.

## Color magnitude diagrams

```python
spec_non_mem = ~np.logical_and(red_cat["z"] < 0.2, red_cat["z"] > 0.175)
```

```python
spec_non_mem = np.logical_and(spec_non_mem, ~np.isnan(red_cat["z"]))
```

```python
CM_diag = [["g-i", "i"], ["g-r", "r"], ["r-i", "i"]]
```

```python
phot_type = ["iso", "aper", "auto"]
```

```python
for phot in phot_type: 
    red_cat["g-i_" + phot] = red_cat["g" + "_" + phot] - red_cat["i" + "_" + phot] 
    red_cat["g-r_" + phot] = red_cat["g" + "_" + phot] - red_cat["r" + "_" + phot] 
    red_cat["r-i_" + phot] = red_cat["r" + "_" + phot] - red_cat["i" + "_" + phot] 
```

```python
spec_non_mem1 = np.logical_and(~np.isnan(red_cat["z"]), red_cat["z"] < 0.175)
spec_non_mem2 = np.logical_and(~np.isnan(red_cat["z"]), red_cat["z"] > 0.2)
spec_non_mem = np.logical_or(spec_non_mem1, spec_non_mem2)
```

```python
np.sum(spec_non_mem)
```

```python
for diag in CM_diag: 
    for phot in phot_type:
        xaxis = diag[1] + "_" + phot
        yaxis = diag[0] + "_" + phot

        plt.plot(red_cat[xaxis][mem_mask], 
            red_cat[yaxis][mem_mask], 'r.', alpha=0.8, label='spec members')
        plt.plot(red_cat[xaxis][spec_non_mem], 
            red_cat[yaxis][spec_non_mem], 'g.', alpha=0.8, label='spec non members')

        plt.plot(red_cat[xaxis], 
            red_cat[yaxis], 'b.', alpha=0.1)
        plt.xlabel(xaxis)
        plt.ylabel(yaxis)
        plt.ylim((0, 2))
        plt.legend(loc="best")
        plt.show()
        plt.close()
        
```
