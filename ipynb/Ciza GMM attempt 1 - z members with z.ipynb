```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn.mixture import GMM 
import itertools
from scipy import linalg
from sklearn.preprocessing import StandardScaler
```

Want to get an idea of what the data set is about

```python
z_cat = pd.read_hdf("../data/CizaClusterMembers.h5", 
                      "RedshiftSelection")
```

Let's try the GMM with a small number of variables first ... namely 

```python
var = ["x_image", "y_image", "z" ]# , "g-i_iso", "r-i_iso", "g-r_iso"] # , "z"]
```

```python
color_iter = itertools.cycle(['k', 'r', 'g', 'b', 'c', 'm', 'y'])
```

```python
X = np.array(z_cat[var])
```

I am doing a bad job of including the redshift info, I standardize all the spatial coordinates with redshift 

```python
X = StandardScaler().fit_transform(X) 
```

```python
n_components_range = range(1, 6)
```

```python
cv_type = "full" # we can try the rest later ["spherical", "tied", "diag", "full"]  
```

# See if GMM can find the smaller spatial subclusters 

```python
bic = []
lowest_bic = np.infty 
# most of this code can be found from scikit learn tutorial 
for n_components in n_components_range:
    gmm = GMM(n_components=n_components, covariance_type=cv_type)
    gmm.fit(X)
    bic.append(gmm.bic(X))
    if bic[-1] < lowest_bic:
            lowest_bic = bic[-1]
            best_gmm = gmm
    clf = gmm
    Y_ = gmm.predict(X)
    splot = plt.subplot(1, 1, 1)
    for i, (mean, covar, color) in enumerate(zip(clf.means_, clf.covars_,
                                             color_iter)):
        v, w = linalg.eigh(covar)  # not sure why the eigenvectors blew up ... 
        if not np.any(Y_ == i):
            continue
        plt.scatter(X[Y_ == i, 0], X[Y_ == i, 1], 2, color=color)
        plt.plot(clf.means_[i][0], clf.means_[i][1], 'x', mew=2,
                 color=color)
        plt.xlabel("x_image")
        plt.ylabel("y_image")
        plt.title("number of cluster = {0}".format(i + 1))
        # Plot an ellipse to show the Gaussian component
        angle = np.arctan2(w[0][1], w[0][0])
        angle = 180 * angle / np.pi  # convert to degrees
        # try different multiplicative factor for the size of the ellipse
        v *= 4  
        ell = mpl.patches.Ellipse(mean, v[0], v[1], 180 + angle, 
                              color=color)
        ell.set_clip_box(splot.bbox)
        ell.set_alpha(.5)
        splot.add_artist(ell)
    plt.show()
    plt.close()

```
